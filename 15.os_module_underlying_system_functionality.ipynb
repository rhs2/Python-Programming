```python
#look some info about files
os.stat("2.integer.ipynb")
```

```python
os.stat("2.integer.ipynb").st_size #bytes
```

```python
#last modification time
os.stat("2.integer.ipynb").st_mtime
```

```python
#modify that st_time as human readable 
import datetime
mod_time = os.stat("2.integer.ipynb").st_mtime
print(datetime.datetime.fromtimestamp(mod_time))
```
