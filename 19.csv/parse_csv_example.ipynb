```python
#get the list into html and check how many are the contibutors are there
import csv
html_output = " "
names = []
with open ("patrons.csv" ,'r') as f:
    content= csv.reader(f)
    #we do not want header first line of bad data

    next(content) #skip header
    next(content) #skip first data line

    for line in content:
        if line[0]== 'No Reward':
            break 
        names.append(f"{line[0]} {line[1]}")
html_output += f" <p> There are currently {len(names)} public contibutors. Thanks to you1 </p> "

html_output += "\n<ul>"

for name in names:
    html_output += f"\n\t <li>{name}</li>"
    
html_output += "\n</ul>"

print(html_output)
```

```python
#using DictReader Method
#get the list into html and check how many are the contibutors are there
import csv
html_output = " "
names = []
with open ("patrons.csv" ,'r') as f:
    content= csv.DictReader(f)
    #we do not want header first line of bad data

    next(content) #skip header
    next(content) #skip first data line

    for line in content:
        if line['FirstName']== 'No Reward':
            break 
        names.append(f"{line['FirstName']} {line['LastName']}")
html_output += f" <p> There are currently {len(names)} public contibutors. Thanks to you </p> "

html_output += "\n<ul>"

for name in names:
    html_output += f"\n\t <li>{name}</li>"
    
html_output += "\n</ul>"

print(html_output)
```
