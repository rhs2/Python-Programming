```python
#choices method it will get some value from a list, getting multiple value
list1 = ['black', 'red', 'white']
value6 = random.choices(list1, k =10) #parameter k will give 10 random value in a list
print(value6)
```

```python
#we also can tell in the arguments how many % we want in the random choices list so we can weight this 
value7 = random.choices(list1, weights=[18,18,2], k =10) #black has 18 out of 36, red has 18 out of 36 and white has 2 out of 36
print(value7)
```

```python
deck = list(range(1,53)) #deck of cards
#using random module to shuffle the list and each time will shuffle the cards 
random.shuffle(deck)
print(deck)
```

```python
#we can also use choices method to shuffle the cards, in choices it can grab the same card , we only want unique cards so we use different method called sample
deck = list(range(1,53))
hand = random.sample(deck, k =5) #because we want 5 different card and unique, we never get same card twice
print(hand)
```
